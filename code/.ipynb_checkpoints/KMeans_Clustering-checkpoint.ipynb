## Third: Let's now describe a function that implements K-Means clustering

It should have the following inputs:
<ul>
    <li> <i>K</i>: K
    <li> <i>data</i>: Array of dimensions (N,D)  | N = Number of data-points; D = Dimensionality of each data-point
    <li> <i>distance</i>: distance calculation function | A function that returns distance between two data-points Eg. Euclidean, Manhattan
    <li> <i>loss</i>: loss measurement function | A function that can evaluate the clustering Eg. TotalVariance, MaxVariance
    <li> <i>I</i>: Iterations
</ul>
It should return:
<ul>
    <li> <i>centroids</i>: Array of dimension (K,D) | This array gives the co-ordinates of each of the cluster-centroids
    <li> <i>associations</i>: Array of dimension (N) | data[i] belongs to cluster with centroid at centroids[associations[i]]  
    <li> <i>loss</i>: best loss after I iterations
        
        

## Fourth: Before we define the K-Means function, Let's define a couple other things


<b><u><i>distance</b></i></u>: Let's implement Euclidean and Manhattan distances <br/>
These functions should be fairly simple. They should take 2 data-points and return the distance betwen them<br/>
<b><u><i>loss</b></i></u>: Let's implement a loss function. We will use the notion of Loss function described by Luis Serrano in his video available at <a href = "https://www.youtube.com/watch?v=QXOkPvFM6NU&t=10m6s"> https://www.youtube.com/watch?v=QXOkPvFM6NU&t=10m6s </a>  <br/>
Loss functions may be complicated. For this particular loss, we only need to provide it data-points, their associations and applicable distance algorithm. centroid co-ordinates are NOT needed. Some other losses may require more or less information to be provided. However, they should all return one numeric value denoting the "goodness" of the clustering


<img src = "..\images\Fig2.jpg" title = "Fig2: Loss Method">

```python
# Implement Distance Function. 

# Let's implement a function for computing Euclidean Distance

def euclidean(a, b):
    return np.sqrt(np.sum(np.power(np.subtract(a,b), 2)))

# Let's also implement a function for computing Manhattan Distance

def manhattan(a, b):
    return np.sum(np.abs(np.subtract(a,b)))
```

```python
# Implement Loss Function.

# Let's implement the loss function described in Fig2; 
# Let's call it LuisSerranoLoss. Because Why NOT? :)

def LuisSerranoLoss(associations, data, distance):
    result = 0
    
    for i in range(data.shape[0]):
        for k in range(data.shape[0]):
            if associations[i] == associations[k]:
                candidate = distance(data[i], data[k])
                result = max(candidate, result)
                
    return result
```

```python
# Let's us test all the functions we have defined so far.

print("\nEuclidean distance between points (1,1) and (2,2) should be ~1.41, \
        our implementation says it is:",
      str(round(euclidean([1,1], [2,2]), 2)))

print("\nManhattan distance between points (1,1) and (2,2) should be 2, \
        our implementation says it is:",
      str(manhattan([1,1], [2,2])))

print("\nFor points:([0, 0], [0,1], [10, 10], [10, 12]) with cluster associations: \
        ([0, 0, 1, 1]),")
print("Most distant points who are in the same cluster are [10, 10] and [10, 12]")
print("The euclidean distance between them is 2; This should be the LuisSerranoLoss as well")
print("According to our implementation, LuisSerranoLoss = ", str(LuisSerranoLoss([0, 0, 1, 1],
        np.array([[0, 0], [0,1], [10, 10], [10, 12]], dtype = np.uint8), euclidean)))
```

## Finally, Let's implement our K-Means function!

### Implementation Notes:

<ul>
<li> Initially, we have to randomly select centroid-points. We just pick K random data-points to act as centroids. Alternatively, we could use completely random K-points in the vector space but that would be a bit more complex and should't be beneficial in any way.
<li> We say that a paricular iteration has converged ONLY when the centroids STOP moving across successive iterations. 

```python
def KMeans(K, data, distance, loss, I=10):
    
#     Let's initialize the variables we plan to return
    best_loss = None
    best_centroids = None
    best_associations = None
    
    
#     As K-Means is unstable, we run it I times and return results of the best run.
    for i in range(I):
        
#         First step is to select K random centroids. As mentioned in implementation notes, 
#         we pick K data-points to act as initial centroids
        selected_indices = []

        while len(selected_indices) != K:
            selected_indx = random.randint(0, data.shape[0] -1)
            if selected_indx not in selected_indices:
                selected_indices.append(selected_indx)

#         Veriables for this particular run
        associations = np.zeros((data.shape[0]), np.uint8)
        curr_loss = None
#         Two array to keep track of centroid co-ordinates pre and post an iteration
        centroids_post = np.array([data[index] for index in selected_indices], dtype = np.float)
        centroids_pre = np.zeros_like(centroids_post)
        
#         Stopping condition: NO CHANGE in ANY centroid co-ordinate after an iteration
        while(not (centroids_pre == centroids_post).all()):
            
#             before an iteration, pre(n) = post(n-1) 
            centroids_pre = np.copy(centroids_post)
            
#             First, we associate each data-point to one centroid
            for k in range(data.shape[0]):
                closest_centroid_indx = 0  
                closest_centroid_distance  = distance(data[k], centroids_pre[0]) 
                for j in range(1, centroids_pre.shape[0]):
                    candidate_distance = distance(data[k], centroids_pre[j])
                    if closest_centroid_distance > candidate_distance:
                        closest_centroid_indx = j
                        closest_centroid_distance = candidate_distance
                        associations[k] = j
                        
#             Then, we move each centroid to the center of all data-points associated with it
            for k in range(centroids_pre.shape[0]):
                data_points_collected = 0
                data_points_coords_collected = np.zeros((data.shape[1]), np.float)
                
                for j in range(0, data.shape[0]):
                    if associations[j] == k:
                        data_points_collected += 1
                        data_points_coords_collected += data[j]
                
                centroids_post[k] = data_points_coords_collected / data_points_collected
#                 There is no need for division-by-zero protection here,
#                 because atleast one data-point is guranteed to stay associated 
#                 to each centroid since the beginning
            
#         We get loss for this particular run of KMeans
        curr_loss = loss(associations, data, distance)
        
#         curr_loss is best loss iff this is the first run or curr_loss < best_loss
#         if curr_loss < best_loss, we save the results
        if i == 0 or best_loss > curr_loss:
            best_loss = curr_loss
            best_associations = np.copy(associations)
            best_centroids = np.copy(centroids_post)
            if best_loss == 0:
                break
#     return only the best results
    return (best_centroids, best_associations, best_loss)
    
```

```python
# Let's do a test run of our K-Means implementation on Toy-data

centroids, associations, loss = KMeans(2, np.array([[0, 0], [0,1], [10, 10], [10, 12]],
                                        dtype = np.uint8), euclidean, LuisSerranoLoss, 10)

print("\nFor points:([0, 0], [0,1], [10, 10], [10, 12]):")
print("Best cluster associations are ([0, 0, 1, 1]), \
        Our implementation says:", str(associations))

print("\nMost distant points who are in the same cluster are [10, 10] and [10, 12]")
print("The euclidean distance between them is 2; This should be the LuisSerranoLoss as well")
print("According to our implementation, LuisSerranoLoss = ", str(loss))

print("\n According to our implementation, best centroids are:\n", centroids)
```

### Now, that we have verified our implementation on Toy-data, Let's run it on <code>faithul.csv</code> data

```python
# Let's transform our data by a little bit.
eruptions = np.expand_dims(data["eruptions"], axis = -1)
waiting = np.expand_dims(data["waiting"], axis = -1)

dataset = np.concatenate((eruptions, waiting), axis = -1)

# And then we call our method with K=2, 
# Let's just print out the centroids so we can do a sanity check
centroids, associations, loss = KMeans(2, dataset, euclidean, LuisSerranoLoss)
print(centroids)
```

The centroids look about right!

## Now, Let's plot the elbow-curve to find the value of K

```python
# cell-parameter to set range of K from [1, try_upto-1]
try_upto = 10

# Let's only accumulate the losses for each value of K
losses = []
for i in range(1, try_upto):
    _, _, loss = KMeans(i, dataset, euclidean, LuisSerranoLoss)
    losses.append(loss)
    

# Let's setup a few labels in our figure
plt.figure(figsize = (5,5))
plt.title("Fig3: Old Faithful Elbow-Graph")
plt.plot([k for k in range(1, try_upto)], losses, label = "LuisSerranoLoss")
plt.xticks([k for k in range(0, try_upto)])
plt.legend()
# render, show and save the plot
plt.savefig(os.path.join("..","images","Fig3.jpg"))
plt.show()
```

### Observation:
From Fig3, It seems that K =2 is correct for this particular dataset.

## Let's run our implementation of KMeans on this data

```python
plt.figure(figsize = (5,5))
plt.title("Fig4: Old Faithful Results")

# Run KMeans with K=2
centroids, associations, loss = KMeans(2, dataset, euclidean, LuisSerranoLoss, I=25)

# Plot the centroids as *
plt.plot(centroids[0][0],centroids[0][1], "r*", label = "centroid1")
plt.plot(centroids[1][0],centroids[1][1], "b*", label = "centroid2")


# Plot all the data-points as per centroid-colors
for i in [0, 1]:
    points_x = []
    points_y = []
    
    for k in range(associations.shape[0]):
        if associations[k] == i:
            points_x.append(data["eruptions"][k])
            points_y.append(data["waiting"][k])
    
    if i==0:
        plt.plot(points_x, points_y, "r.", label = "cluster1")
    elif i==1:
        plt.plot(points_x, points_y, "b.", label = "cluster2")

# Add labels
plt.xlabel("eruptions")
plt.ylabel("waiting")
plt.legend()

# render, show and save the graph
plt.savefig(os.path.join("..","images","Fig4.jpg"))
plt.show()
```
